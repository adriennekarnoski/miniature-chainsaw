## Getting Started

### Configure **matplotlib inline** to show figures embedded in the notebook and display plot as a **SVG**:

`%...`: magic functions in iPython

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

### Import **numpy** and **matplotlib.pyplot** module:

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Set up your variables:

```python
pi = np.pi
x = np.linspace(-4*pi, 4*pi, 1000)
```

### Call **plot** with args and **show** the figure:

`plot(*args, **kwargs)`: plot lines and/or markers to the axes

`show()`: display figure and return to the ipython prompt

```python
plt.plot(x, np.sin(x)/x)
plt.show()
```

Set tick labels:

```python
plt.plot(x, np.sin(x)/x)
plt.yticks([0, 1], ['$0$', '$1$'])
plt.xticks([-4*pi, -2*pi, 0, 2*pi, 4*pi],
          ['$-4\pi$', '$-2\pi$', '$0$', '$2\pi$', '$4\pi$'])
plt.ylim(-0.5, 1.5)
```

Adjust figure size:

```python
plt.plot(x, np.sin(x)/x)
plt.rcParams['figure.figsize'] = (11, 4)
plt.rcParams.update({'font.size': 17})
```

Create a legend and title:
- `$...$` **LaTeX notation** for inline mathematics
- `\begin{equation}...\end{equation}` for standard equations

```python
plt.plot(x, np.sin(x)/x, label=r'$f(x)=\frac{\sin(x)}{x}$')
plt.title('The $sinc$ Function')
plt.legend(loc='best', fontsize=20)
plt.axhline(0, color='black', lw=1)
plt.show()
```
